## 1. Два основных варианта применения

1. **Критерий согласия (goodness of fit)** — сравниваем **наблюдаемые частоты** категорий с **теоретическим распределением** (например, равные доли при случайном угадывании в эксперименте с 4 альтернативами).

2. **Критерий независимости** — есть **две номинальные переменные**; проверяем, **связаны ли они** в выборке (например, **тип психообразования** и **готовность обратиться за помощью**).

В этом ноутбуке основной упор на **критерий независимости** в таблице сопряжённости (**contingency table**), так как такие задачи очень типичны в прикладной психологии.

## 2. Идея критерия (без лишней математики)

Мы сравниваем, **насколько сильно «реальные» числа в ячейках** отличаются от чисел, которые **ожидались бы**, если бы переменные были **полностью независимы** (только структура маргиналов — итогов по строкам и столбцам — оставалась бы той же).

**Статистика Пирсона:**

$$\chi^2 = \sum \frac{(O_{ij} - E_{ij})^2}{E_{ij}}$$

где $O_{ij}$ — наблюдаемая частота в ячейке $(i,j)$, $E_{ij}$ — ожидаемая частота при гипотезе независимости.

**Ожидаемая частота** для ячейки:

$$E_{ij} = \frac{(\text{сумма строки } i) \times (\text{сумма столбца } j)}{\text{общее } N}$$

**Степени свободы** для таблицы $r \times c$:

$$df = (r-1)(c-1)$$

## 3. Гипотезы (формулировки для отчёта)

- **$H_0$:** переменные **независимы** в генеральной совокупности (в популяции нет связи; отклонения в выборке — случайны).  
- **$H_1$:** переменные **зависимы** (связь есть).

Если **$p < \alpha$** (часто $\alpha = 0{,}05$), **отвергаем $H_0$** и говорим, что связь **статистически значима** (при соблюдении условий применимости).

## 4. Условия применимости (важно для защиты курсовой)

1. **Номинальные** (или порядковые, если вы сознательно группируете) категории; наблюдения **независимы** (обычно разные респонденты).  
2. Правило частот: чаще всего требуют, чтобы **ожидаемные** частоты $E_{ij}$ были **не меньше 5** (иногда допускают до 20% ячеек с $E$ от 1 до 5 — зависит от учебника/журнала; в сомнительных случаях используют **точный** тест Фишера для $2\times2$).  
3. Критерий $\chi^2$ чувствителен к **объёму выборки**: при очень большом $N$ даже **слабые** связи могут стать «значимыми» — смотрите **размер эффекта** (ниже — **V Крамера**).


## 5. Психологический пример исследования (выдуманные данные)

**Исследовательский вопрос:** связан ли **уровень социальной тревожности** (ниже / выше медианы по опроснику) с **предпочтением формата психологической помощи** (очная консультация vs онлайн)?

**Дизайн:** репрезентативная для учебных целей выборка студентов, $N=120$. Две номинальные переменные:

- **Строки:** тревожность — «Низкая», «Высокая» (разделение по медиане — учебный приём; в реальной работе обосновывают пороги).  
- **Столбцы:** предпочтение — «Очно», «Онлайн».

**Таблица наблюдаемых частот:**

|                    | Очно | Онлайн | Всего (строка) |
|--------------------|-----:|-------:|---------------:|
| Низкая тревожность |  35  |   15   |      50        |
| Высокая тревожность|  20  |   50   |      70        |
| **Всего (столбец)**|  55  |   65   |     120        |

Мы проверим независимость этих двух признаков при $\alpha = 0{,}05$.


## 6. Пошаговые ручные расчёты ожидаемых частот и $\chi^2$

**Обозначения:** $R_i$ — сумма по строке $i$, $C_j$ — сумма по столбцу $j$, $N=120$.

**Ячейка (Низкая, Очно):** $E_{11} = \frac{50 \times 55}{120} = \frac{2750}{120} \approx 22{,}92$  

**Ячейка (Низкая, Онлайн):** $E_{12} = \frac{50 \times 65}{120} \approx 27{,}08$  

**Ячейка (Высокая, Очно):** $E_{21} = \frac{70 \times 55}{120} \approx 32{,}08$  

**Ячейка (Высокая, Онлайн):** $E_{22} = \frac{70 \times 65}{120} \approx 37{,}92$

**Вклад каждой ячейки** в $\chi^2$: $\frac{(O-E)^2}{E}$

| Ячейка   | $O$  | $E$     | $(O-E)^2$ | $(O-E)^2/E$ |
|----------|-----:|--------:|----------:|------------:|
| Низк–Очно| 35   | 22,917  | 145,694   | 6,357       |
| Низк–Онл.| 15   | 27,083  | 146,007   | 5,391       |
| Выс–Очно | 20   | 32,083  | 146,007   | 4,551       |
| Выс–Онл. | 50   | 37,917  | 146,007   | 3,856       |

**Сумма:** $\chi^2 \approx 6{,}357 + 5{,}391 + 4{,}551 + 3{,}856 \approx 20{,}15$ (уточним в коде).

**Степени свободы:** таблица $2\times2$ → $df = (2-1)(2-1) = 1$.

Далее сравним полученное $\chi^2$ с **критическим** значением из распределения $\chi^2$ при $df=1$ и $\alpha=0{,}05$, а также вычислим **точное $p$-значение** в Python.


## 7. Таблица критических значений $\chi^2$ (фрагмент для ориентира)

Ниже — **верхние критические значения**: если **наблюдаемое** $\chi^2$ **больше** табличного при выбранных $df$ и $\alpha$, область **отклонения $H_0$** (при классическом одностороннем сравнении для этого теста).

| $df$ | $\alpha=0{,}10$ | $\alpha=0{,}05$ | $\alpha=0{,}01$ | $\alpha=0{,}001$ |
|-----:|----------------:|----------------:|-----------------:|------------------:|
| 1    | 2,706           | **3,841**       | 6,635            | 10,828            |
| 2    | 4,605           | 5,991           | 9,210            | 13,816            |
| 3    | 6,251           | 7,815           | 11,345           | 16,266            |
| 4    | 7,779           | 9,488           | 13,277           | 18,467            |
| 5    | 9,236           | 11,070          | 15,086           | 20,515            |

Для нашего примера: $df=1$, $\alpha=0{,}05$ → критическое значение **3,841**. Если $\chi^2 > 3{,}841$, отвергаем $H_0$.

*Примечание:* точные значения удобно получать функцией квантиля (в коде ниже) — таблица дана для понимания «бумажной» процедуры.


````python
# Импорты: NumPy — массивы и суммы; scipy.stats — готовый тест и распределение хи-квадрат
import numpy as np
from scipy import stats

# --- Исходная таблица наблюдаемых частот (строки: низкая/высокая тревожность; столбцы: очно/онлайн) ---
# Важно: порядок строк и столбцов нужно сохранять одинаковым во всех выводах и интерпретации.
observed = np.array(
    [
        [35, 15],  # низкая тревожность
        [20, 50],  # высокая тревожность
    ],
    dtype=float,
)

row_labels = ["Низкая тревожность", "Высокая тревожность"]
col_labels = ["Очно", "Онлайн"]

print("Наблюдаемые частоты O:")
print(observed)
print()
````

````python
def expected_frequencies(contingency: np.ndarray) -> np.ndarray:
    '''Ожидаемые частоты при H0 независимости: E_ij = (сумма строки i)*(сумма столбца j)/N.'''
    row_sums = contingency.sum(axis=1, keepdims=True)  # суммы по строкам
    col_sums = contingency.sum(axis=0, keepdims=True)  # суммы по столбцам
    total = contingency.sum()
    # Внешнее произведение маргиналов / N даёт матрицу E той же формы, что и O
    expected = row_sums @ col_sums / total
    return expected


def chi_square_pearson(observed, expected):
    '''Статистика Пирсона: sum(O-E)^2/E. Возвращает (chi2, вклады по ячейкам).'''
    # Избегаем деления на ноль: в реальных данных при нулевой E нужна другая стратегия (слияние категорий и т.д.)
    contributions = (observed - expected) ** 2 / expected
    chi2 = float(contributions.sum())
    return chi2, contributions


E = expected_frequencies(observed)
chi2_manual, contrib = chi_square_pearson(observed, E)

n_rows, n_cols = observed.shape
df = (n_rows - 1) * (n_cols - 1)
N = observed.sum()

print("Ожидаемые частоты E (при H0 — независимость):")
print(np.round(E, 3))
print()
print("Вклад каждой ячейки (O-E)^2/E:")
for i in range(n_rows):
    for j in range(n_cols):
        print(
            f"  {row_labels[i]} × {col_labels[j]}: "
            f"O={observed[i,j]:.0f}, E={E[i,j]:.3f}, вклад={contrib[i,j]:.4f}"
        )
print()
print(f"χ² (Пирсон, вручную по формуле) = {chi2_manual:.4f}")
print(f"Степени свободы df = {df}")
print(f"Объём выборки N = {N:.0f}")
````

````python
# Сравнение с критическим значением из распределения χ² (тот же результат, что в статистических таблицах)
alpha = 0.05
chi2_crit = stats.chi2.ppf(1 - alpha, df)  # верхний квантиль: P(χ² > crit) = alpha
p_value_upper = stats.chi2.sf(chi2_manual, df)  # sf = 1 - cdf; для стандартного теста — именно «правый хвост»

print(f"Уровень значимости α = {alpha}")
print(f"Критическое значение χ²_crit(df={df}, α={alpha}) = {chi2_crit:.4f}")
print(f"p-значение (правый хвост) P(χ² ≥ {chi2_manual:.4f}) = {p_value_upper:.6e}")
print()
if chi2_manual > chi2_crit:
    print("Вывод по сравнению с табличным критерием: χ² наблюдаемое > χ² критическое → отвергаем H0.")
else:
    print("Вывод: χ² наблюдаемое ≤ χ² критическое → нет оснований отвергать H0.")
````

````python
# Проверка через scipy.stats.chi2_contingency (должно совпасть с ручным χ² для этой задачи)
chi2_scipy, p_scipy, dof_scipy, expected_scipy = stats.chi2_contingency(observed)

print("scipy.stats.chi2_contingency:")
print(f"  χ² = {chi2_scipy:.6f}")
print(f"  p  = {p_scipy:.6e}")
print(f"  df = {dof_scipy}")
print("  Ожидаемые частоты (должны совпасть с E выше):")
print(np.round(expected_scipy, 3))
assert np.allclose(chi2_manual, chi2_scipy, rtol=1e-9, atol=1e-9)
assert np.allclose(E, expected_scipy)
print("\nПроверка: ручной χ² совпал с chi2_contingency.")
````

````python
def cramers_v(contingency: np.ndarray) -> float:
    '''V Крамера: sqrt(chi2/(N*min(r-1,c-1))). Ориентиры ~0.1 / 0.3 / 0.5 (грубо, осторожно для r,c>2).'''
    chi2 = stats.chi2_contingency(contingency, correction=False)[0]
    n = contingency.sum()
    r, c = contingency.shape
    k = min(r - 1, c - 1)
    if k <= 0:
        return float("nan")
    return float(np.sqrt(chi2 / (n * k)))


v = cramers_v(observed)
print(f"V Крамера = {v:.4f}")
print("(Интерпретация: насколько сильна связь между признаками, помимо «значимо/не значимо».)")
````

````python
# Расширенная таблица критических значений χ² для df = 1..10 и нескольких α (как справочник для студентов)
alphas = [0.10, 0.05, 0.025, 0.01, 0.001]
df_list = list(range(1, 11))

col_w = 12
header = "df".ljust(4) + "".join(str(a).center(col_w) for a in alphas)
print("Критические значения χ² (верхний хвост): строка = df, столбцы = α")
print(header)
print("-" * len(header))
for d in df_list:
    parts = [str(d).ljust(4)]
    for a in alphas:
        crit = stats.chi2.ppf(1 - a, d)
        parts.append(f"{crit:>{col_w}.3f}")
    print("".join(parts))
````

## 8. Интерпретация для психологического отчёта (выводы исследования)

**Статистика:** для выдуманных данных получается **большое** значение $\chi^2$ при $df=1$ и **очень малое** $p$-значение (проверьте число в выводе ячейки кода выше). Это означает: **если бы** в генеральной совокупности предпочтение формата и уровень социальной тревожности были **полностью независимы**, то таблица такого вида встречалась бы **крайне редко** только из-за случайной выборочной ошибки.

**Содержательный вывод (осторожно!):** статистика **не говорит причину**. Мы можем сказать: **связь есть** между **группой по тревожности** и **предпочтением очного vs онлайн формата** в этой (учебной) выборке. Возможные **объяснения** (для обсуждения, не для «доказательства» из одного $\chi^2$): доступность, стигма, привычка к очным контактам, самоэффикасия в digital-среде и т.д. — это уже теория и последующие исследования.

**Размер эффекта:** сообщайте **V Крамера** рядом с $\chi^2$ и $p$, чтобы читатель видел **силу** связи, а не только значимость при большом $N$.

**Оформление в стиле APA (пример шаблона):**  
«Связь между уровнем социальной тревожности и предпочтением формата консультации оказалась значимой, $\chi^2(1, N=120) = \ldots$, $p < .001$, V Крамера $= \ldots$.»

**Ограничения:** выдуманные частоты; медианный сплит снижает мощность и спорен методологически; для малых ожидаемых частот смотрите **точный тест Фишера** (`scipy.stats.fisher_exact` для $2\times2$).


````python
# Дополнительно: точный тест Фишера для 2×2 (когда ожидаемые частоты малы — альтернатива χ²)
# Для нашей таблицы ожидаемые частоты > 5, Fisher здесь скорее иллюстрация инструмента.
odds_ratio, p_fisher = stats.fisher_exact(observed)
print(f"Тест Фишера (2×2): odds ratio = {odds_ratio:.4f}, p = {p_fisher:.6e}")
print("Интерпретация OR: во сколько раз отличаются шансы предпочесть один столбец в первой строке относительно второй (с осторожностью к причинности).")
````
